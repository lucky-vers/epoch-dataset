functions

```python
#master binary : use this for binary columns, there are many
#pass the column name and then a tuple containing the custom values which is optional, by default it is no/yes

def master_binary(column_name: str, custom_text=('no', 'yes')):
    df[column_name] = df[column_name].map({0: custom_text[0], 1: custom_text[1], np.nan: 'N/A'}) 
```

```python
def changecol1():
    df['AGE'] = df['AGE'].apply(lambda x: str(int(x)) + 'y 0m' if not pd.isna(x) else 'N/A')

def changecol2():
    df['SEX'] = df['SEX'].map({0: 'female', 1: 'male'})

def changecol3():
    df['INF_ANAM'] = df['INF_ANAM'].map({ 0: 'zero', 1: 'one', 2: 'two', 3:'three'})

def changecol4():
    df['STENOK_AN'] = df['STENOK_AN'].map({0: 'never', 1: 'last summer', 2:'1y ago', 3:'2y ago', 4:'3y ago', 5:'4-5y ago',6:'>5y ago', np.nan: 'N/A'})

def changecol5():
    df['FK_STENOK'] = df['FK_STENOK'].map({0:'none', 1:'I FC', 2: 'II FC', 3: 'III FC', 4: 'IV FC'})

def changecol6():
    df['IBS_POST'] = df['IBS_POST'].map({0: 'none', 1: 'exertional', 2: 'unstable'})

def changecol8():
    df['GB'] = df['GB'].map({0:'none', 1: 'Stage 1', 2: 'Stage 2', 3: 'Stage 3', np.nan: 'N/A'})

def changecol10():
    df['DLIT_AG'] = df['DLIT_AG'].map({0:'none', 1: '1y', 2:'2y', 3:'3y', 4:'4y', 5:'5y', 6:'6-10y', 7:'>10y', np.nan: 'N/A'})

def changecol11():
    df['ZSN_A'] = df['ZSN_A'].map({0:'none', 1: 'I stage', 2: 'II stage',  3: 'II stage',  4: 'IIB stage', np.nan: 'N/A'})
```

applying transformations in data

```python
changecol1()
changecol2()
changecol3()
changecol4()
changecol5()
changecol6()
master_binary('IBS_NASL', ('burdened', 'not burdened')) #7
changecol8()
master_binary('SIM_GIPERT') #9
changecol10()
changecol11()

#12 - 22

remaining_binary_12_22 = ['nr_11', 'nr_01', 'nr_02', 'nr_03', 'nr_04', 'nr_07', 'nr_08', 'np_01', 'np_04', 'np_05', 'np_07']
for feature in remaining_binary_12_22:
    master_binary(feature)

```
